```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('EPL 2000-2018.csv', index_col=0, parse_dates=True)

data.head()
```

```python
data['Year'] = data.index.year
data['Month'] = data.index.month_name()
data['Weekday Name'] = data.index.weekday_name
# Display a random sampling of 5 rows
data.sample(5, random_state=0)
```

```python
sns.set(rc={'figure.figsize':(9, 5)})
```

```python
plt.scatter(data['HC'],data['FTHG'])
plt.xlabel("HC", fontsize=20)
plt.ylabel("FTHG", fontsize=20)
plt.show()
```

```python
data['FTR'] = data['FTR'].map({'H':0,'D':1,'A':2})
data.head()
```

```python
data['HTR'] = data['HTR'].map({'H':0,'D':1,'A':2})
```

## FTR CLUSER ANALYSIS

```python
kmeans = KMeans(3)

kmeans.fit(x_clean)
```

```python
identified_clusters = kmeans.fit_predict(x_clean)
identified_clusters
```

```python
data_with_clusters = x_clean.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters.head()
```

```python
plt.scatter(data_with_clusters['HC'],data_with_clusters['FTHG'],c=data_with_clusters['Cluster'],cmap='rainbow')

plt.show()
```
